```python
import wikipediaapi
import wikipedia
import concurrent.futures

```

```python

# Set the language to English
wiki = wikipediaapi.Wikipedia('en')

def get_random_wikipedia_page():
    page = None
    while page is None or not page.exists():
        try:
            random_title = wikipedia.random(pages=1)
            page = wiki.page(random_title)
        except Exception as e:
            print(f"Exception occurred while downloading article: {e}")
    return page

```

```python

def download_wikipedia_article(i):
    page = get_random_wikipedia_page()
    print(f"Downloading article {i}: {page.title}")
    article_data = {
        "title": page.title,
        "text": page.text
    }
    return article_data

# Set the number of articles to download (change this to your desired number of articles)
num_articles = 5000
num_threads = 40

articles_data = {}

with concurrent.futures.ThreadPoolExecutor(max_workers=num_threads) as executor:
    future_to_i = {executor.submit(download_wikipedia_article, i): i for i in range(num_articles)}
    for future in concurrent.futures.as_completed(future_to_i):
        i = future_to_i[future]
        try:
            data = future.result()
            articles_data[f"article_{i}"] = data
        except Exception as e:
            print(f"Exception {e} occurred while downloading article {i}")

# Write articles data to file
with open("articles.txt", "w", encoding="utf-8") as f:
    for article_id, article_data in articles_data.items():
        f.write(f"{article_id}\n{article_data['title']}\n{article_data['text']}\n\n")

print("Data extraction completed.")
```
