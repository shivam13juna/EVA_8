```python
# source_dir = "dataset/source/"
# target_dir = "dataset/target/"
# Define source and target directories
source_dir = "image_dataset"
target_dir = "dataset/target"
canny_dir = "dataset/source"


# Create the target and canny directories if they don't exist
if not os.path.exists(target_dir):
    os.makedirs(target_dir)
if not os.path.exists(canny_dir):
    os.makedirs(canny_dir)

# Initialize an empty list to store the image data
image_data = []

# Get a list of all images in the source directory
image_files = os.listdir(source_dir)

progress_bar = tqdm(total=len(image_files), unit="image")


# Loop over each image file
for i, image_file in enumerate(image_files, start=1):
    # Define the source path
    source_path = os.path.join(source_dir, image_file)
    
    # Define the target path
    target_path = os.path.join(target_dir, f"{i}.jpg")
    
    
    
    # Read the image using OpenCV
    img = cv2.imread(source_path, cv2.IMREAD_GRAYSCALE)
    
    # Generate the Canny edge detection image
    edges = cv2.Canny(img, 100, 200)
    
    # Define the canny image path
    canny_path = os.path.join(canny_dir, f"{i}.jpg")
    try:
        # Write the edges image to the source directory
        cv2.imwrite(canny_path, edges)

        # Copy the image to the target directory
        shutil.copy2(source_path, target_path)
        
        # Add the image data to the list
        image_data.append({
            "source": canny_path,
            "target": target_path,
            "prompt": image_file.replace(".jpg", "").replace("__", " ")
        })

        # Update the progress bar
        
    except:
        pass

    progress_bar.update(1)
# Close the progress bar
progress_bar.close()

# Write the image data to a JSON file
with open("image_data.json", "w") as outfile:
    for data in image_data:
        json.dump(data, outfile)
        outfile.write("\n")
```
