```python
import os
from PIL import Image

# Define the path to your image directory
image_dir = 'dataset/target/'

# List to store the sizes
image_sizes = []

# Variables to store the smallest width and height
smallest_width = float('inf')
smallest_height = float('inf')

# Iterate over the images in the directory
for filename in os.listdir(image_dir):
    # Check if the file is an image
    if filename.endswith('.jpg') or filename.endswith('.png'):  # add more conditions here if you have different image types
        # Open the image
        try:
            img = Image.open(os.path.join(image_dir, filename))
        except:
            print(f"Error loading image: {filename}")
            continue

        # Get the size
        width, height = img.size

        # Append the size to the list
        image_sizes.append((width, height))

        # Check if the width is the smallest
        if width < smallest_width:
            smallest_width = width

        # Check if the height is the smallest
        if height < smallest_height:
            smallest_height = height

        if width < 100:
            os.remove(os.path.join(image_dir, filename))
            os.remove(os.path.join('dataset/source/', filename))

# The smallest width and height
smallest_size = (smallest_width, smallest_height)

print(f"All image sizes: {image_sizes}")
print(f"Smallest image size: {smallest_size}")
```
